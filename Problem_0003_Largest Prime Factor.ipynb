```python
n = 600851475143
```

```python
factors = []
```

```python
from sympy import nextprime
```

```python

check_factor = nextprime(0)
```

```python
while n > 1:
    print("=================================")
    print("n = ", n)
    print("check_factor = ", check_factor)
    if n % check_factor == 0:
        factors.append(check_factor)
        n = int(n / check_factor)

    else:
        check_factor = nextprime(check_factor)
    
```

```python
max(factors)
```
